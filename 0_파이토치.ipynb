### 1-4. 텐서(Tensor)
- 다수의 행렬이 모이면 텐서
- 배열이나 행렬과 매우 유사한 특수 자료구조
- 파이토치는 텐서를 사용하여 모델의 입력과 출력, 모델의 매개변수들을 처리 사용

<img src = 'https://miro.medium.com/max/875/1*jRyyMAhS_NZxqyv3EoLCvg.png' width = 500>

```python
A = torch.tensor([[1, 2, 3], [4, 5, 6]])
B = torch.tensor([10, 20, 30])

# 브로드캐스팅을 사용하여 A와 B를 덧셈
# 작은 배열이 큰 배열로 확장되어 연산을 수행했다.
print(A)
print('=======================')
print(B)
print('=======================')
print(A + B)
```

```python
tensor1 = torch.tensor([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print(tensor1)
```

```python
tensor2 = torch.tensor([[[10, 11], [12, 13]], [[14, 15], [16, 17]]])
print(tensor2)
```

```python
print(tensor1 + tensor2)
print('==============================')
print(tensor1 - tensor2)
print('==============================')
print(tensor1 * tensor2)
print('==============================')
print(tensor1 / tensor2)
```

```python
print(torch.add(tensor1, tensor2))
print('==============================')
print(torch.subtract(tensor1, tensor2))
print('==============================')
print(torch.multiply(tensor1, tensor2))
print('==============================')
print(torch.divide(tensor1, tensor2))
```

```python
# tensor1에 결과를 다시 저장, 모든 사칙연산자에 _를 붙여 사용할 수 있음.
print(tensor1.add_(tensor2))
```

# **2. 텐서의 변환**

```python
data = [[1, 2], [3, 4]]
data
```

```python
x_data = torch.tensor(data)
x_data
```

```python
import numpy as np
```

```python
np_array = np.array(data)
np_array
```

```python
x_np_1 = torch.tensor(np_array)
x_np_1
```

```python
# 인덱싱? okay
x_np_1[0, 0] = 100
print(x_np_1)
```

```python
# as_tensor : ndarray와 동일한 메모리 주소를 가리키는 뷰를 만듦
# from_numpy : ndarray와 동일한 메모리 주소를 가리키는 뷰를 만듦
x_np_2 = torch.as_tensor(np_array)
print(x_np_2)
x_np_2[0, 0] = 200
print(x_np_2)
print(np_array)

# 반대로도 확인!
print('===============================')
np_array[0, 0] = 500
print(x_np_2)
print(np_array)
```

```python
np_again = x_np_1.numpy()
print(np_again, type(np_again))
```

# **5. 코랩의 GPU 사용**
- 코랩에서 device 변경하는 방법
  - 상단 메뉴 -> 런타임 -> 런타임 유형변경 -> 하드웨어 가속기를 GPU로 변경 -> 저장 -> 다시 시작 및 모두 실행

```python
tensor = torch.rand(3, 4)
print(f'shape -> {tensor.shape}')
print(f'shape -> {tensor.dtype}')
print(f'shape -> {tensor.device}')
```

```python
tensor = torch.rand(4, 3)
# 데이터의 형변환!
tensor = tensor.int()

if torch.cuda.is_available():
  print('GPU 사용 가능')
  tensor = tensor.to('cuda')

print(f'shape -> {tensor.shape}')
print(f'shape -> {tensor.dtype}')
print(f'shape -> {tensor.device}')
```
